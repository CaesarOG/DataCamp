<h1>Matplotlib</h1>
Used for data visualization, a very important part of data analysis. Used continuously to explore data. The better you understand the data, the better you can extract insights. Need visualization again to share insights with others. By making detailed beautiful plots, let data tell its own story. Like when graphing per capita GDP as x and Life Expectancy as y for some countries, each data point's size on the graph is determined by the country's population. Points also color coded for region of the world. Sizing and coloring of points helps data speak for itself. Not a perfect positive corr, some countries with same per capita GDP have very disparate expectancies, and good graph helps figure out why.

<h1>Dictionaries & Pandas</h1>

````python
populations = [30.55, 2.77, 39.21]
countries = ["Afghanistan", "Albania", "Algeria"]
ind_alb = countries.index("Albania") 
populations[ind_alb] #have to know index of Albania first and only then can
#use fact that same indexes to get Albania's population.
````

````python
#dictionaries are the efficient convenient way to connect info like country to its population.
world = {"Afghanistan": 30.55, "Albania":2.77, "Algeria":39.21}
world = {} #Now a efficient way of above.
for i in range(len(countries)):
    world[countries[i]] = populations[i]

world['Albania']#key value pairs. 
````

````python
world.keys() #shows the keys in the dictionary.
````

For lookup to work properly, keys must be unique. Try to initialize a dictionary with the 
same key two diff values, the last value is kept.
```python
world = {"Afghanistan": 30.55, "Albania":2.77, "Algeria": 39.21, "Albania":2.81}
```
Above, last val for Albania, 2.81, overwrites val for that key. Keys have to be immutable objects, can't change a key only del/make new ones. Number, bool, string can all be keys but list can't since can change contents.

````python
world["Sealand"] = 0.000027 #added new country to world.
world["Sealand"] = 0.000028 #knows update and not new since unique keys.
"Sealand" in world
````

````python
del(world["Sealand"]) #deleted it.
world #printout not necessarily same order as definition, dictionaries inherently unordered.
````

List and dictionary work is similar. In both, you select update and remove elts with square brackets \[\]. List is indexed by range of numbers (int), and dictionary on other hand indexed by unique keys. Use list when order matters and want to easily select entire subsets of data. Use dictionary when need lookup table for unique keys.

````python
#just as list-within-list possible, dictionary-within-dictionary.
europe = { 'Spain': { 'capital': 'Madrid', 'population': 46.77}, 'France': { 'capital': 'Paris',
    'population': 66.03
}}
data = {'capital': 'Rome', 'population': 59.83}
europe['Italy'] = data
europe['Italy']['capital']
````

Form of data varies could be image/audio/video, but often its structured tabular data, we use pandas for that. Lets say temp measurements, each row is observations/samples the columns are the temp, measured_at time, and location. Or let's say have BRIC countries' capital, area and population and name, those are the columns and each row is another observation/sample.
2d Numpy array is option but not best one, they can only house elts with same data type, but just in BRIC example saw area and population were floats and capital and country are string. Pandas dataframe lets columns be different types. Datasets will typically comprise different data types. Pandas is data manip tool built by Wes McKinney, built on Numpy but more high level. In the dataframe version of BRIC data, still have observations as rows and columns as are the attributes of each sample/observations.

````python
import pandas as pd

#different ways to make dataframe, can manual from dictionary.
dict1 = { #could assign each list to var like countries and do "country": countries
    "country": ["Brazil", "Russia", "India", "China", "South Africa"], 
    "capital": ["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
    "area": [8.516, 17.10, 3.286, 9.597, 1.221],
    "population": [200.4, 143.5, 1252, 1357, 52.98]
} #keys are col labels, values are data col by col.
brics = pd.DataFrame(dict1)
````

````python
brics["country"] #select one col from dataframe, returns pandas Series. 
brics[["country"]] #still one col, but use two sets of [] to get back a 1-col dataframe.

brics[["country", "capital"]] #sub dataframw with those two cols.

#Just as a Dataframe is like 2d np array w/ labelling, Series like 1d np array w/ labelling.
````

````python
brics[1:4] #standard int slicing for rows, 2nd->5th exclusive so rows 2,3,4 

#with np arrays, can do arr[rows, cols] where both are either int index or slice of int index.
#loc and iloc do this in pandas. loc is label-based, iloc is int based w/ python int slicing.

brics.loc["RU"] #a series, use [[]] two sets for dataframe.
#print(y[x == "Z"]) bool array of which indices in x are "Z", used to select same indices in y.
````

````python
brics.loc[:, ["country", "capital"]] #all rows those two cols.
brics.loc[["RU", "IN"], :] #all cols those two rows.
brics.loc[['RU'], ['capital']] #one value.
brics.loc[["RU", "IN", "CH"], ["country", "capital"]] #multiple rows and cols by label.
````

Simple square brackets get column access like `brics[["country", "capital"]]` , or row access (only int slicing) like `brics[1:4]` . Loc more versatile can select only by rows `brics.loc[["RU", "IN", "CH"]]` or only by cols `brics.loc[:, ["country", "capital"]]` . Or both rows and cols `brics.loc[["RU", "IN", "CH"], ["country", "capital"]]` .

````python
brics.iloc[[1]] #same as brics.loc[["RU"]] .
brics.iloc[[1,2,3]] #same as brics.loc[["RU", "IN", "CH"]] .

brics.iloc[[1,2,3], [0,1]] #same as brics.loc[["RU", "IN", "CH"], ["country", "capital"]] .

brics.loc[:, 'country'] == brics.iloc[:, 0] #true
brics.iloc[1:4,0:2] #same output as above, full int subslicing not just : like in loc.

brics.iloc[:, 0:2] #or iloc[:, [0,1]] . Both same as brics.loc[:, ["country", "capital"]] .
````

<h1>Logic, Control Flow and Filtering</h1>

Remember bmi numpy array from first course, bmi > 23 returned bool np array, true for each elt if it is > 23. Then use these bools to subset get all where that's true, `bmi[bmi > 23]` . Behind the scenes for boolean subset, Numpy made array of same size with all elts = 23 and did elt-wise comparison. Comparison operators like > tell you how Python values relate.

````python
"true" <= "true" #true
2 < 3 #true
2 == 3 #false
True > False #true
"carl" < "chris" #true, alphabetical order for strings
x, y = 2,3
x != y #true
True >= 1 #coercion, rare, should mostly compare numbers or same exact type.
````

````python
my_house = np.array([18.0, 20.0, 10.75, 9.50]) 
your_house = np.array([14.0, 24.0, 14.25, 9.0])
my_house < your_house #boolean subset example two arrays.
````

````python
#boolean and.
True and True #only true when both true
False and True #false
True and False #false
False and False #false

x = 12
x > 5 and x < 15 #both true in and, so true
````

````python
#boolean or.
True or False #true
False or True #true
True or True #true
False or False # or only false when both false

y = 5
y < 7 or y > 13 #first part true, so true
````

````python
#boolean not.
not True #false
x = 8
y = 9
not(not(x < 3) and not(y > 14 or y > 10)) #false
````

````python
#what if want to 'and'/'or' numpy boolean subsets? ordinary && || wont work need special
#logic operators.
bmi = np.array([21.852, 20.975, 21.75, 24.747, 21.441])
#can't do bmi > 21 and bmi < 22.
print(bmi[np.logical_and(bmi > 21, bmi < 22)])
np.logical_and(bmi > 21, bmi < 22) #the two boolean arrays are 'and' together.
````

````python
print(np.logical_or(my_house > 18.5, my_house < 10))
print(np.logical_and(my_house < 11, your_house < 11))
````

````python
z = 6 #Control Structure. breaks entire control structure after 1st if matched, never reaches other ifs like %3 (also true).
if z % 2 == 0: #if statement, takes expression that returns boolean
    print("z is even, divisible by 2")
elif z % 3 == 0: #elif optional could have just done an else, and say odd.
    print("z is divisible by 3")
else: #no condition, if none of its matching if statements matched else will run.
    print("z neither divisible by 2 or 3")
````

````python
#Filtering. like select * from brics where area > 8; (million) . need series, no double [[]] .
brics[brics["area"] > 8] #inner could also be brics.loc[:, "area"] or brics.iloc[:,2] .
````

````python
#area between 8 and 10.
brics[ np.logical_and( brics["area"] > 8 , brics["area"] < 10 ) ]
````

<h1>Loops</h1>

````python
#while
error = 50.0
while error > 1: #keeps going until error < 1
    error = error / 4 #simulate ML process where error cut into quarter each training epoch.
    print(error) #trace code in your head, how outputs the 3 vals. after 3 runs it's < 1 so done.
````

````python
offset = -6 #Offset! whoo, whoo, whoo, whoo, whoo!
while offset is not 0: #or != 0
    if offset > 0: #Listen! without these conditions -6 gets subtracted forever, infinite loop.
        offset = offset - 1
    else:
        offset = offset + 1
````

````python
#two ways to loop string, notice i is the value itself in first example, because using x but is the index in second example because using range len(x), have to use 
#index into x to get value there.
x = "abc"

for i in x: #for this syntax x must be iterable, behind scenes python calls iter(x), gets iterator.
    print(i) #i is arbitary name, could say var.
    
for i in range(len(x)): #again i is arbitrary name.
    print(x[i])
````

````python
for idx, char in enumerate(x): #gets you both index and elt.
    print("index "+str(idx) +": "+char) #don't have to str(char) already string.
````
