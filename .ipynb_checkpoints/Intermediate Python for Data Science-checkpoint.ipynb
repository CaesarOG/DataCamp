For lookup to work properly, keys must be unique. Try to initialize a dictionary with the 
same key two diff values, the last value is kept.

Above, last val for Albania, 2.81, overwrites val for that key. Keys have to be immutable objects, can't change a key only del/make new ones. Number, bool, string can all be keys but list can't since can change contents.

<h1>Logic, Control Flow and Filtering</h1>

Remember bmi numpy array from first course, bmi > 23 returned bool np array, true for each elt if it is > 23. Then use these bools to subset get all where that's true, `bmi[bmi > 23]` . Behind the scenes for boolean subset, Numpy made array of same size with all elts = 23 and did elt-wise comparison. Comparison operators like > tell you how Python values relate.

````python
#Logic Operators
"true" <= "true" #true
2 < 3 #true
2 == 3 #false
True > False #true
"carl" < "chris" #true, alphabetical order for strings
x, y = 2,3
x != y #true
True >= 1 #coercion, rare, should mostly compare numbers or same exact type.
````

````python
my_house = np.array([18.0, 20.0, 10.75, 9.50]) 
your_house = np.array([14.0, 24.0, 14.25, 9.0])
my_house < your_house #boolean subset example two arrays.
````

````python
#boolean and.
True and True #only true when both true
False and True #false
True and False #false
False and False #false

x = 12
x > 5 and x < 15 #both true in and, so true
````

````python
#boolean or.
True or False #true
False or True #true
True or True #true
False or False # or only false when both false

y = 5
y < 7 or y > 13 #first part true, so true
````

````python
#boolean not.
not True #false
x = 8
y = 9
not(not(x < 3) and not(y > 14 or y > 10)) #false
````

````python
#what if want to 'and'/'or' numpy boolean subsets? ordinary && || wont work need special
#logic operators.
bmi = np.array([21.852, 20.975, 21.75, 24.747, 21.441])
#can't do bmi > 21 and bmi < 22.
print(bmi[np.logical_and(bmi > 21, bmi < 22)])
np.logical_and(bmi > 21, bmi < 22) #the two boolean arrays are 'and' together.
````

````python
print(np.logical_or(my_house > 18.5, my_house < 10))
print(np.logical_and(my_house < 11, your_house < 11))
````

````python
z = 6 #Entire control structure breaks after 1st if matched, never reaches other ifs like 
#%3 (even tho that's also true).

if z % 2 == 0: #if statement, takes expression that returns boolean
    print("z is even, divisible by 2")
elif z % 3 == 0: #elif optional could have just done an else, and say odd.
    print("z is divisible by 3")
else: #no condition, if none of its matching if statements matched else will run.
    print("z neither divisible by 2 or 3")
````

<h1>Case Study: Stats</h1>

Walking up Empire State Building, playing game. Throw dice 100 times, if 1/2 go down one step. If 3/4/5 go up one step. If 6 thrown, throw again and whatever # go up that amt of steps. Can't go below step 0, and 0.1% chance falling down stairs to start at step zero again. You bet you'll get 60 steps high. One way to solve all this is calc chance analytically. Another way is to simulate whole process 1000s of times and see what fraction of sims you reach 60 steps. That is hacker statistics and is what we'll do.

````python
np.random.rand() #num btw 0 and 1, pseudo random: uses a math formula, starting from random
#seed that's generated when Python calls this func. Can also set manually.
np.random.seed(123)
print(np.random.rand())
print(np.random.rand())
np.random.seed(123) #same seed, same numbers
print(np.random.rand())
print(np.random.rand()) 
#pseudo random useful, ensures reproducibility.
````

````python
np.random.seed(123) 
coin = np.random.randint(0,2) #bounds, upper exclusive like in slicing. For dice it's 1,7.
if coin == 0: #coin always 0 or 1.
    print("heads")
else: 
    print("tails") #this was example how to use rand nums to simulate real probab scenarios.
````

The 100x times you throw the dice are the random steps, collectively called random walk. The path of a molecule as it travels through liquid/gas, or financial status of a gambler, are examples of things in science modeled as random walks. 

````python
np.random.seed(123)
outcomes = [] #keep outcomes of coin toss 10 times
for x in range(10) : #range(n) goes 0 to n-1
    coin = np.random.randint(0, 2)
    if coin == 0:
        outcomes.append("heads")
    else :
        outcomes.append("tails")
print(outcomes) #list is random but not random walk, items not based on previous ones.
````

````python
#turn into random walk, by tracking num tails.
np.random.seed(123)
tails = [0]
for x in range(10):
    coin = np.random.randint(0,2)
    tails.append(tails[x] + coin)
print(tails) #1st elt before 1st toss, but shows amt tails after each toss.
````

````python
np.random.seed(123)
random_walk = [0] #first/starting step
for x in range(100):
    step = random_walk[-1] #last elt of walk is current step
    dice = np.random.randint(1,7)
    if dice <= 2:
        step = max(0, step-1) #so don't go below 0
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)
    random_walk.append(step)
print(random_walk)
````

````python
plt.plot(random_walk) #if pass 1 arg, knows to use index of list for x, values as y.
plt.show()
````

Let's talk about distribution of random walks. Depending on the result you go some steps up or down in the example. Still have question, what's chance of 60 steps? Imagine if you repeat the 100x throws a thousand times each walk will end on some step, then you'll have distribution of final steps. Then can calc chance of 60. Think back to tails example, num tails starts zero, and 10 times calc random 0 or 1. Then update num tails by appending to list. 

````python
np.random.seed(123)
final_tails = [] #store tails at end of each 10 throw walk.
for x in range(10000): #amt walks
    tails = [0]
    for x in range(10): #code from before, the 10 toss walk
        coin = np.random.randint(0,2)
        tails.append(tails[x]+coin)
    final_tails.append(tails[-1]) #get final count of each walk before reset walk
plt.hist(final_tails, bins=10)

#hist smoother with 1000 walks vs 100. Changed again to 10000 and starts to have bell shape,
#like a theoretical distribution, very smooth. If you did it millions of times would be exact.
````

````python
#but that's too much compute time. 10000 gives good estimate. Can see that in 2500 of the 10k
#trials, got 5 tails.
np.random.seed(123)
all_walks = []
for i in range(500): #amt walks
    random_walk = [0]
    for x in range(100): #code from before, 100 throw walk
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step-1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= .001 : 
            # 0.1% chance fall down, this syntax works since rand gives float 0 to 1.
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)
#all_walks is list of lists, 10 full walks of 100 throws, each walk showing step progression.
np_aw = np.array(all_walks) 
plt.plot(np_aw) #took row indices of all_walks as x, but rows are the walks, want step as x.
plt.show()
````

````python
np_aw_t = np.transpose(np_aw) #shape goes from (10, 101) to (101, 10)
plt.plot(np_aw_t) 
#All 1st steps from each walk in first row, etc. Now, row indices are the steps because each 
#walk is a column now instead of row.
plt.show()
````

````python
#above can see some walks you did trip and fall back down, few falls quite big & later too.
ends = np_aw_t[-1] #the transpose, so last row is all the last steps
plt.hist(ends)
````

````python
aboveSixty = 0
for i in range(len(ends)):
    if ends[i] >= 60:
        aboveSixty +=1
aboveSixty/500 #for 500 trials have 78.4% chance, finally question answered.
````
