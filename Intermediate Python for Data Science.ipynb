<h4>Matplotlib</h4>
Used for data visualization, a very important part of data analysis. Used continuously to explore data. The better you understand the data, the better you can extract insights. Need visualization again to share insights with others. By making detailed beautiful plots, let data tell its own story. Like when graphing per capita GDP as x and Life Expectancy as y for some countries, each data point's size on the graph is determined by the country's population. Points also color coded for region of the world. Sizing and coloring of points helps data speak for itself. Not a perfect positive corr, some countries with same per capita GDP have very disparate expectancies, and good graph helps figure out why.

```python
from jupyterthemes import jtplot #fun fact if use jupyterthemes, don't have to plt.show() .
jtplot.style()
import matplotlib.pyplot as plt #or from matplotlib import pyplot as plt.
year = [1950, 1972, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972] #in billions.
plt.plot(year, pop) #x,y axes. so want to see pop as a function of year.
plt.show() #always waits to show graph until plt.show(), in case want to add labels, etc.
```

```python
#Python drew line between the four data points. Can scatterplot same data, see below.

plt.scatter(year, pop) #simply plots individual data points. No connecting so can see
#amt of points, often scatter > line; if points not all close to line a line plot looks messy. 
```
