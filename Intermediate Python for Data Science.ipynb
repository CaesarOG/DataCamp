<h4>Dictionaries & Pandas</h4>

````python
populations = [30.55, 2.77, 39.21]
countries = ["Afghanistan", "Albania", "Algeria"]
ind_alb = countries.index("Albania") 
populations[ind_alb] #have to know index of Albania first and only then can
#use fact that same indexes to get Albania's population.
````

````python
#dictionaries are the efficient convenient way to connect info like country to its population.
world = {"Afghanistan": 30.55, "Albania":2.77, "Algeria":39.21}
world = {} #Now a efficient way of above.
for i in range(len(countries)):
    world[countries[i]] = populations[i]

world['Albania']#key value pairs. 
````

````python
world.keys() #shows the keys in the dictionary.
````

For lookup to work properly, keys must be unique. Try to initialize a dictionary with the 
same key two diff values, the last value is kept.
```python
world = {"Afghanistan": 30.55, "Albania":2.77, "Algeria": 39.21, "Albania":2.81}
```
Above, last val for Albania, 2.81, overwrites val for that key. Keys have to be immutable objects, can't change a key only del/make new ones. Number, bool, string can all be keys but list can't since can change contents.

````python
world["Sealand"] = 0.000027 #added new country to world.
world["Sealand"] = 0.000028 #knows update and not new since unique keys.
"Sealand" in world
````

````python
del(world["Sealand"]) #deleted it.
world #order of printout not necessarily same as definition, dictionaries inherently unordered.
````

List and dictionary work is similar. You can select update and remove elts in both with square brackets \[\]. List is indexed by range of numbers (int), and dictionary on other hand indexed by unique keys. Use list when order matters and want to easily select entire subsets of data. Use dictionary when need lookup table for unique keys.

````python
#just as list-within-list possible, dictionary-within-dictionary.
europe = { 'Spain': { 'capital': 'Madrid', 'population': 46.77}, 'France': { 'capital': 'Paris',
    'population': 66.03
}}
europe['France']['capital']
````

Form of data varies could be image/audio/video, but often its structured tabular data, we use pandas for that. Lets say temp measurements, each row is observations/samples the columns are the temp, measured_at time, and location. Or let's say have BRIC countries' capital, area and population and name, those are the columns and each row is another observation/sample.
2d Numpy array is option but not best one, they can only house elts with same data type, but just in BRIC example saw area and population were floats and capital and country are string. Pandas dataframe lets columns be different types. Datasets will typically comprise different data types. Pandas is data manip tool built by Wes McKinney, built on Numpy but more high level. In the dataframe version of BRIC data, still have observations as rows and columns as are the attributes of each sample/observations.

````python
import pandas as pd

#different ways to make dataframe, can manual from dictionary.
dict1 = { 
    "country": ["Brazil", "Russia", "India", "China", "South Africa"],
    "capital": ["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
    "area": [8.516, 17.10, 3.286, 9.597, 1.221],
    "population": [200.4, 143.5, 1252, 1357, 52.98]
} #keys are col labels, values are data col by col.
brics = pd.DataFrame(dict1)
````

````python
brics.index = ["BR", "RU", "IN", "CH", "SA"] #overwrite 0-n int indices.
brics
````
